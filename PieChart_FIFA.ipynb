```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv('Documents/datafiles/NumPan/fifa_data.csv')

fifa
```

```python
# creating a piechart for Preferred foot
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0] 
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]

plt.figure(figsize = (8,6))
# left
# right ---descriptions are plural
labels = ['Left', 'Right']
colors = ['blue', 'red']
explode = (.1,.1)

plt.pie([left, right], labels = labels, colors = colors, explode = explode, autopct ='%.1f %%')
plt.title ("Foot Prefrence of FIFA Players 2018")

plt.show()
```

```python
# a piechart showing the distribution of weight of the players
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
# for removal of the lbs

plt.figure(figsize = (10,8))
plt.style.use ('ggplot')
plt.title('Weight Distribution of FIFA Players')

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc [(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc [(fifa.Weight >= 200)].count()[0]
# fifa.Weight[0]
# light_meduim
weight = [light,light_medium,medium,medium_heavy,heavy]
labels = ['Under 125', '125-150', '150-175','175-200', 'Over 200']
explode = (.4,.1,0,.1,.4)

plt.pie(weight, labels = labels, explode = explode, autopct ='%.2f %%')

# plt.savefig('Pie chart.png', dpi = 200)
plt.show()
```

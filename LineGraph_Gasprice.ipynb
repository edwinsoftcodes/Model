```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas = pd.read_csv ('Documents/datafiles/NumPan/gas_prices.csv')

gas
```

```python
# LINE GRAPH FOR NORTH AMERICA AND ASIA
plt.figure(figsize = (9,5))
plt.title("Gas Prices {in Usd}", fontdict = {'fontweight' : 'bold', 'fontsize' : 22})

plt.plot (gas['Year'], gas.USA, 'b.-', label = 'USA')
plt.plot (gas['Year'], gas.Canada, 'y.-', label = 'Can')
plt.plot (gas['Year'], gas['South Korea'], 'r.-', label = 'Sth K')
plt.plot (gas['Year'], gas.Japan, 'g.-', label = 'JPN')

# for country in gas:
#     if country != 'Year':
#         pt.plot(gas.Year, gas[country], marker = '.')

plt.xlabel('Year')
plt.ylabel('US dollars')
plt.xticks(gas.Year[::3].tolist() + [2011])


# pt.savefig('Gas price figure1.png', dpi = 200)

plt.legend()

plt.show()
```
